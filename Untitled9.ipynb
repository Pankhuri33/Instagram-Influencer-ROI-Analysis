```python
import pandas as pd
import numpy as np
import random
```

```python
# Set seed for consistency
random.seed(42)
np.random.seed(42)
```

```python
# Sample influencer usernames and categories
usernames = [f"@influencer_{i}" for i in range(1, 101)]
categories = ["Skincare", "Makeup", "Haircare", "Fragrance", "Lifestyle"]

data = []

for i in range(100):
    username = usernames[i]
    category = random.choice(categories)
    followers = random.randint(10000, 1000000)
    posts = random.randint(1, 10)
    likes = random.randint(500, 20000)
    comments = random.randint(50, 1500)
    reach = random.randint(int(0.4 * followers), followers)
    cost = round(random.uniform(1000, 100000), 2)
    revenue = round(random.uniform(cost * 0.5, cost * 2), 2)  # realistic variation

    data.append([username, category, followers, posts, likes, comments, reach, cost, revenue])

# Create DataFrame
df = pd.DataFrame(data, columns=[
    "Influencer", "Category", "Followers", "Posts", "Likes", "Comments",
    "Reach", "Campaign_Cost", "Revenue"
])
```

```python
# Add derived metrics
df["Engagement_Rate"] = ((df["Likes"] + df["Comments"]) / df["Followers"]) * 100
df["CPM"] = df["Campaign_Cost"] / (df["Reach"] / 1000)
df["ROI"] = (df["Revenue"] - df["Campaign_Cost"]) / df["Campaign_Cost"]

# Save to CSV
df.to_csv("influencer_campaign_data.csv", index=False)

# Preview
print(df.head())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("influencer_campaign_data.csv")

# Show first 5 rows
print(df.head())

# Basic info
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Check for null values
print(df.isnull().sum())

# Check for duplicates
print("Duplicate rows:", df.duplicated().sum())
```

```python
# Top 5 Influencers by ROI
top_roi = df.sort_values(by="ROI", ascending=False).head(5)
print("Top 5 by ROI:")
print(top_roi[["Influencer", "ROI", "Revenue", "Campaign_Cost"]])

# Best Engagement Rate
top_engagement = df.sort_values(by="Engagement_Rate", ascending=False).head(5)
print("Top 5 by Engagement Rate:")
print(top_engagement[["Influencer", "Engagement_Rate", "Likes", "Comments"]])

# Most Cost-effective (Lowest CPM)
lowest_cpm = df.sort_values(by="CPM").head(5)
print("Top 5 by lowest CPM:")
print(lowest_cpm[["Influencer", "CPM", "Reach", "Campaign_Cost"]])
```

```python
import matplotlib.pyplot as plt

# ROI of Top 10 Influencers
top10_roi = df.sort_values("ROI", ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.bar(top10_roi["Influencer"], top10_roi["ROI"], color='mediumseagreen')
plt.xticks(rotation=45, ha='right')
plt.title("Top 10 Influencers by ROI")
plt.ylabel("ROI")
plt.tight_layout()
plt.savefig("top10_roi.png")
plt.show()

# Engagement vs Cost
plt.figure(figsize=(8,6))
plt.scatter(df["Engagement_Rate"], df["Campaign_Cost"], color='tomato')
plt.title("Engagement Rate vs Campaign Cost")
plt.xlabel("Engagement Rate (%)")
plt.ylabel("Campaign Cost (INR)")
plt.grid(True)
plt.tight_layout()
plt.savefig("engagement_vs_cost.png")
plt.show()

# CPM Distribution
plt.hist(df["CPM"], bins=20, color='cornflowerblue')
plt.title("Distribution of CPM (Cost per 1000 impressions)")
plt.xlabel("CPM")
plt.ylabel("Frequency")
plt.savefig("cpm_distribution.png")
plt.show()
```

```python
def categorize_follower_group(followers):
    if followers < 100000:
        return "Micro"
    elif 100000 <= followers <= 500000:
        return "Mid"
    else:
        return "Macro"

df["Influencer_Type"] = df["Followers"].apply(categorize_follower_group)
```

```python
roi_by_type = df.groupby("Influencer_Type")["ROI"].mean().reset_index()
print(roi_by_type)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(x="Influencer_Type", y="ROI", data=roi_by_type, palette="Set2")
plt.title("Average ROI by Influencer Type")
plt.ylabel("ROI")
plt.xlabel("Influencer Type")
plt.tight_layout()
plt.savefig("roi_by_type.png")
plt.show()
```
